## INSTAGRAM STATISTICS AND DATA VISUALIZATION

```python

#Change Username value and run the code

#string
USERNAME = '_alexander_osipenko_'

from InstaData import *

data = get_all_posts(USERNAME)
data = reshape_for_plotting(data)
```

```python

import scipy
import scipy.misc
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.dates as mdates
from matplotlib import style
style.use('ggplot')


mu = round(scipy.mean(data['likes']))  # mean of distribution
sigma = round(np.std(data['likes']))  # standard deviation of distribution
fig, ax = plt.subplots()
num_bins = 50
n, bins, patches = ax.hist(data['likes'], num_bins, normed=1)
# add a 'best fit' line
y = mlab.normpdf(bins, mu, sigma)
ax.plot(bins, y, '--', )
ax.set_xlabel('Likes')
ax.set_ylabel('Probability density')
ax.set_title(r'Histogram of Likes: $\mu=' + str(mu) + '$, $\sigma=' + str(sigma) +  '$')
# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()

fig, ax1 = plt.subplots()
ax1.plot_date(data['created_time'], data['likes'])
ax1.tick_params(axis='both', which='major', labelsize=7)
ax1.set_ylabel('Likes')
ax1.set_title(r'Max Likes ' + str(np.amax(data['likes']))+ ' On ' + str(data['created_time'][np.argmax(data['likes'])]) )
fig.tight_layout()
plt.show()
```
